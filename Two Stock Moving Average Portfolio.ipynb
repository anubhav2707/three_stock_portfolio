# CALCULATING RETURNS

```python
total_m=an.TOTAL.resample('M').mean()
nifty_m=an['NIFTY 50'].resample('M').mean()
```

```python
#function to calculate percentage return
def returns(col,s,name):
    for i in reversed(range(len(s))):
        b=(s.iloc[i][col]-s.iloc[i-1][col])/(s.iloc[i-1][col])
        s.iloc[i, s.columns.get_loc(name)]=b
        
def only_returns(col,s,name):
    for i in reversed(range(len(s))):
        b=(s.iloc[i][col]-s.iloc[i-1][col])
        s.iloc[i, s.columns.get_loc(name)]=b
        
```

```python
#storing monthly return in returnm
temp1= pd.DataFrame(total_m)
temp2=pd.DataFrame(nifty_m)
returnm=pd.concat([temp1,temp2],axis='columns')
returns(0,returnm,"TOTAL")
returns(1,returnm,"NIFTY 50")
returnp=pd.concat([temp1,temp2],axis='columns')
only_returns(0,returnp,"TOTAL")
only_returns(0,returnp,"NIFTY 50")
returnm
```

```python
#Storing yearly returns in returny
total_y=an.TOTAL.resample('Y').mean()
nifty_y=an['NIFTY 50'].resample('Y').mean()
temp1=pd.DataFrame(total_y)
temp2=pd.DataFrame(nifty_y)

returny=pd.concat([temp1,temp2],axis='columns')
returns(0,returny,"TOTAL")
returns(1,returny,"NIFTY 50")

returny
```

# SHARPES RATIO

```python
def sharpe(data,stock,index):
    data["sharpe"]=0
    std1=data[data.columns[stock]].std()
    for i in range(1,len(data)):
        data.iloc[i, data.columns.get_loc('sharpe')] = (data.iloc[i][stock]-data.iloc[i][index])/std1
        
          
    
```

```python
sharpe(returnm,0,1)
sharpe(returny,0,1)
returnm
```

## Calculating the Drawdown

```python
def dow(s,name,start,end):
    s.reset_index(inplace=True)
    mask = s[(s['Date'] >= start) & (s['Date'] <= end)]
    
    temp=mask.copy()
    temp1=mask.copy()
    temp["cumulative"]=0
    temp["HighValue"]=0

    for i in range(len(mask)):
        temp['cumulative'] = temp[name].cumsum().round(2)
        temp['HighValue'] = temp['cumulative'].cummax()
        temp['Drawdown'] = temp['cumulative'] - temp['HighValue']
    print("Max Drawdown :")
    print(temp['Drawdown'].min())
    return temp
```
